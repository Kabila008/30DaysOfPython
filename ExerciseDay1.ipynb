
 Find the Euclidian distance between (2,3) and (10,8)<br>

    Formular = squareroot((x2 - x1)**2 + (y2 - y1)**2)<br>
    math.sqrt or d**0.5<br>


```python
d = (((10 - 2)**2 + (8 - 3)**2)**0.5)
print('The Euclidian distance is {}.'.format(d))
```
