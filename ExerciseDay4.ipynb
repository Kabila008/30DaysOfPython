    36. Make the following using string formatting methods:<br>
    8 + 6 = 14<br>
    8 - 6 = 2<br>
    8 * 6 = 48<br>
    8 / 6 = 1.33<br>
    8 % 6 = 2<br>
    8 // 6 = 1<br>
    8 ** 6 = 262144<br>


```python
a = 8
b = 6
```

```python
print('{} + {} = {}'.format(a, b, a + b))
print('{} - {} = {}'.format(a, b, a - b))
print('{} * {} = {}'.format(a, b, a * b))
print('{} / {} = {:.2f}'.format(a, b, a / b))
print('{} % {} = {}'.format(a, b, a % b))
print('{} // {} = {}'.format(a, b, a // b))
print('{} ** {} = {}'.format(a, b, a ** b))
```

```python
a = 8
b = 6
```

```python
print(f'{a} + {b} = {a + b}')
print(f'{a} - {b} = {a - b}')
print(f'{a} * {b} = {a * b}')
print(f'{a} / {b} = {a / b : .2f}')
print(f'{a} % {b} = {a % b}')
print(f'{a} // {b} = {a // b}')
print(f'{a} ** {b} = {a ** b}')
```
